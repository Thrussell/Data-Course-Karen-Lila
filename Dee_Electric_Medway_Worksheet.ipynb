Multiple data in bar charts.  Can't just change the bar type to bar (Well you can, but it won't cope with multiple sets of data, need numpy for that).


```python
import numpy as np
import matplotlib.pyplot as plt
"""
multiple lines - like in line chart
Need to use numpy as there are multiple lines.  
Use width to set a skinny bar (rather than 0.8)
Don't forget to add width to the plt.bar so it knows how wide to do the bar (or they don't overlap)
use width to tell it where to place each coloured bar x_indexes - width for example will put it to the left by 0.25
Ages not shown correctly, its using indexes, so add xtick so that it uses the x_age information correctly
"""

x_age = [25, 26, 27, 28, 29, 30]

x_indexes = np.arange(len(x_age))

width = 0.25
y_salary = [23500, 30000, 35000, 45000, 47000, 60000]
y_pyth_salary = [24000, 31000, 37000, 46000, 47500, 62000]
plt.bar(x_indexes-width,y_salary, color="g", width=width,label="Any Developer")
plt.bar(x_indexes,y_pyth_salary, color="b", width=width, label="Python Developer")

plt.xticks(ticks=x_indexes, labels=x_age)
plt.title("Average Developer Salaries by Age")
plt.ylabel('Average Salaries')
plt.xlabel('Age')
plt.legend()
plt.show()
```
